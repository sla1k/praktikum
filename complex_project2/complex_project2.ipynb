### Проверка правильности расчёта эффективности обогащения

Проверим правильность расчёта эффективности обогащения `rougher.output.recovery` на обучающей выборке, предварительно очистив данные от пропуска, т.к. нам нужно проверить расчёт целевого признака, сейчас можем просто удалить эти данные, содержащие 0 и NaN. 

```python
data_train_clean = data_train.dropna(subset=['rougher.output.concentrate_au', 'rougher.input.feed_au', 'rougher.output.tail_au', 'rougher.output.recovery'])
data_train_clean.isnull().sum()
```

```python
data_train_clean = data_train_clean[(data_train_clean['rougher.output.concentrate_au'] != 0) 
                 & (data_train_clean['rougher.input.feed_au'] != 0) 
                 & (data_train_clean['rougher.output.tail_au'] != 0)]
data_train_clean.head()
```

```python
def calculate_recovery_after_rough(row):
    rougher_output_concentrate_au = row['rougher.output.concentrate_au']
    rougher_input_feed_au = row['rougher.input.feed_au']
    rougher_output_tail_au = row['rougher.output.tail_au']
    
    return (rougher_output_concentrate_au * (rougher_input_feed_au - rougher_output_tail_au)) / (rougher_input_feed_au * (rougher_output_concentrate_au - rougher_output_tail_au)) * 100
    
```

```python
recovery_after_rough = data_train_clean['rougher.output.recovery']
calculated_recovery_after_rough = data_train_clean.apply(calculate_recovery_after_rough, axis=1)
```

```python
mean_absolute_error(calculated_recovery_after_rough, recovery_after_rough)
```

Среднее абсолютное отклонение у нас очень мало, из чего делаем вывод, что расчёт `rougher.output.recovery` происходит верно.

## Построение модели

Перед тем как тренировать модель напишем функцию расчёта sMAPE.

```python
def smape(target, predict):
    return 100 / len(target) * np.sum(np.abs(target - predict) / ((np.abs(predict) + np.abs(target)) / 2))
```

```python
def calculate_total_smape(rougher_smape, final_smape):
    return 0.25 * rougher_smape + 0.75 * final_smape
```

Разделим наши обучающую и тестовую выборки на признаки и целевые признаки.

```python
scaller = StandardScaler()
```

```python
X_train = data_train.drop(columns=['rougher.output.recovery', 'final.output.recovery'], axis=1)
y_train_rough = data_train['rougher.output.recovery']
y_train_final = data_train['final.output.recovery']

X_train = scaller.fit_transform(X_train)
```

```python
X_test = data_test.drop(columns=['rougher.output.recovery', 'final.output.recovery'], axis=1)
y_test_rough = data_test['rougher.output.recovery']
y_test_final = data_test['final.output.recovery']

X_test = scaller.transform(X_test)
```

```python
smape_scorer = make_scorer(smape, greater_is_better=False)
```

```python
forest_param_grid = {'max_depth': [deep for deep in range(4, 12, 2)],
                    'n_estimators': [50, 100, 1000]}
```

```python
forest_rough_svr = GridSearchCV(RandomForestRegressor(),
                   scoring=smape_scorer,
                   cv=5,
                   param_grid=forest_param_grid, n_jobs = -2)

forest_rough_svr.fit(X_train, y_train_rough)
```

```python
print (forest_rough_svr.best_params_)
print ('Forest train sMAPE rough: {}'.format(forest_rough_svr.score(X_train, y_train_rough)))
forest_rough_smape = forest_rough_svr.score(X_test, y_test_rough)
print ('Forest test sMAPE rough: {}'.format(forest_rough_smape))
```

```python
forest_final_svr = GridSearchCV(RandomForestRegressor(),
                   scoring=smape_scorer,
                   cv=5,
                   param_grid=forest_param_grid, n_jobs = -2)

forest_final_svr.fit(X_train, y_train_final)
```

```python
print (forest_final_svr.best_params_)
print ('Forest train sMAPE final: {}'.format(forest_final_svr.score(X_train, y_train_final)))
forest_final_smape = forest_final_svr.score(X_test, y_test_final)
print ('Forest test sMAPE final: {}'.format(forest_final_smape))
```

```python
linear_param_grid = {'normalize': [True, False]}
```

```python
linear_rough_svr = GridSearchCV(LinearRegression(), param_grid=linear_param_grid,
                   scoring=smape_scorer,
                   cv=5, n_jobs = -2)

linear_rough_svr.fit(X_train, y_train_rough)
```

```python
print (linear_rough_svr.best_params_)
print ('Linear train sMAPE rough: {}'.format(linear_rough_svr.score(X_train, y_train_rough)))
linear_rough_smape = linear_rough_svr.score(X_test, y_test_rough)
print ('Linear test sMAPE rough: {}'.format(linear_rough_smape))
```

```python
linear_final_svr = GridSearchCV(LinearRegression(), param_grid=linear_param_grid,
                   scoring=smape_scorer,
                   cv=5, n_jobs = -2)

linear_final_svr.fit(X_train, y_train_final)
```

```python
print (linear_final_svr.best_params_)
print ('Linear train sMAPE final: {}'.format(linear_final_svr.score(X_train, y_train_final)))
linear_final_smape = linear_final_svr.score(X_test, y_test_final)
print ('Linear test sMAPE final: {}'.format(linear_final_smape))
```

Результаты получились не однозначными, для чернового концетрата более точным оказалась модель случайного леса, а для финального результата лучше оказалась линейная регрессия. Проверим наши модели на адекватность. Из-за особенности make_score score получился отрицательными, можно не обращать внимание на '-'.

```python
dummy_regr_rough = DummyRegressor(strategy="mean")
dummy_regr_rough.fit(X_train, y_train_rough)

dummy_rough_smape = smape(y_test_rough, dummy_regr_rough.predict(X_test))
print('Dummy smape rough: {}'.format(dummy_rough_smape))
```

```python
dummy_regr_final = DummyRegressor(strategy="mean")
dummy_regr_final.fit(X_train, y_train_final)
dummy_final_smape = smape(y_test_final, dummy_regr_final.predict(X_test))
print('Dummy smape final: {}'.format(dummy_final_smape))
```

```python
dummy_total_smape = calculate_total_smape(dummy_rough_smape, dummy_final_smape)
print('Итоговой dummy sMAPE: {}'.format(dummy_total_smape))
```

Наши модели хоть и немного, но лучше предсказывают чем константная модель, значит она адекватна и эффективна.

Посчитаем итоговый sMAPE.
