Так же сделаем масштабирование данных для лучшего обучения логистической регрессии.

```python
numeric_columns = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']

def scaling(data):
    scaller = StandardScaler()
    scaller.fit(data[numeric_columns])

    scaled_data = data.copy()
    scaled_data.loc[:, numeric_columns] = scaller.transform(scaled_data.loc[:, numeric_columns])
    return scaled_data

X_scaled_train = scaling(X_train)
X_scaled_valid = scaling(X_valid)
X_scaled_test = scaling(X_test)
```

## Обучение моделей

```python
def print_auc_roc(model, X, y):
    probabilities_valid = model.predict_proba(X)
    probabilities_one_valid = probabilities_valid[:, 1]

    auc_roc = roc_auc_score(y, probabilities_one_valid)

    print('AUC-ROC: {}'.format(auc_roc))
```

Так как дерево решений это частный случай частного леса, то остановимся на двух моделях - случайный лес и логистической регресии. Подбор гиперпараметров оставим доверим GridSearchCV. Начнём со случайного леса.

```python
forest_param_grid = {'max_depth': [deep for deep in range(4, 16, 2)],
                    'n_estimators': [50, 100, 150]}
forest_gs = GridSearchCV(RandomForestClassifier(random_state=123), param_grid=forest_param_grid)
forest_gs.fit(X_train.drop(ohencoder.categories_[0], axis=1), y_train)

print(forest_gs.best_params_)
```

На основе текущих данных лучшими гиперпараметрами оказались глубина дерева - 12, а количество деревьев - 100.

```python
forest_predict = forest_gs.predict(X_valid.drop(ohencoder.categories_[0], axis=1))

print(classification_report(y_valid, forest_predict))
print('F1-score: {}'.format(f1_score(y_valid, forest_predict)))
print_auc_roc(forest_gs, X_valid.drop(ohencoder.categories_[0], axis=1), y_valid)
print()
print('Train score: {}'.format(forest_gs.score(X_train.drop(ohencoder.categories_[0], axis=1), y_train)))
print('Valid score: {}'.format(forest_gs.score(X_valid.drop(ohencoder.categories_[0], axis=1), y_valid)))
```

С помощью classification_report мы можем сразу обратить внимание на все интерисующие нас метрики. Мы можем наблюдать достаточно высокий accuracy на валидационной выборке, но мы видим что у нас есть дисбаланс классов, 0 предсказывается гораздо лучше чем 1, об этом на говрит метрика F1-score. ## Балансировка классов

Так как классы целевого признака не сбалансированы, мы попробуем сделать балансировку с помощью техник Up Sempling и Down Sempling.

```python
ex_1_index = y_train[y_train == 1].index
ex_0_index = y_train[y_train == 0].index

ex_1 = data.loc[ex_1_index]
ex_0 = data.loc[ex_0_index]

ex_0.shape[0], ex_1.shape[0]
```

```python
difference = ex_0.shape[0] - ex_1.shape[0]
difference
```

Попробуем наростить количество данных со значением класса 1 применив UpSampling.

```python
up_X = ex_1.sample(difference, replace=True).drop('Exited', axis=1)
up_y = np.repeat(1, difference)

X_upsampling_train = pd.concat([X_train, up_X])
y_upsampling_train = np.hstack([y_train.values, up_y])
X_upsampling_train.shape, y_upsampling_train.shape
```

```python
np.unique(y_upsampling_train, return_counts=True)
```

```python
X_upsampling_train.head()
```

Попробуем уменьшить количество данных со значением класса 0 применив DownSampling.

```python
down_X = ex_0.sample(ex_1.shape[0]).drop(['Exited'], axis=1)
down_y = np.repeat(0, down_X.shape[0])

X_downsampling_train = pd.concat([down_X, ex_1.drop('Exited', axis=1)])
y_downsampling_train = np.hstack((down_y, y_train[y_train != 0].values))
X_downsampling_train.shape, y_downsampling_train.shape
```

```python
np.unique(y_downsampling_train, return_counts=True)
```

```python
X_downsampling_train.head()
```

Мы выровняли значения классов двумя способами.
